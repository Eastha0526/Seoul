# Data Load

- 서울시 킥보드 수거 데이터에서의 수거 구와 유형 그리고 날짜, 서울시 유동인구 데이터에서 날짜, 유동 인구수 등 분석에 필요한 컬럼만 추출했다.

```python
board = pd.read_csv("./data/seouldata.csv", index_col="index", encoding = "cp949")
```

```python
human = pd.read_csv("./data/human.csv", encoding = "cp949")
```

```python
human.columns
```

```python
human_df = human[["기준일ID", "시군구명","주간인구수(09~18)", "야간인구수(19~08)", "일최대이동인구수"]]
```

```python
human_df.head()
```

```python
board["Date"] = board["신고일"].apply(lambda x : x.split("-"))
```

```python
board["Date"] = board["Date"].apply(lambda x : x[0]+x[1]+x[2]).astype("int64")
```

```python
board_df = board[["Date", "구정보", "유형"]]
```

# Viusalization

```python
max_human = human_df.groupby("시군구명").mean()
```

```python
max_human = max_human.drop("서울시")
```

```python
plt.figure()
sns.scatterplot(data = max_human, x = "일최대이동인구수", y = max_human.index)
plt.show()
```

- 일최대 이동 인구수를 비교해 봤을 때, 강남구가 가장 많고 그 뒤를 서초구, 양천구가 그 뒤를 따랐다.

```python
fig, ax = plt.subplots(nrows=2, figsize = (16, 16))

sns.scatterplot(data = max_human, x = "주간인구수(09~18)", y = max_human.index, ax = ax[0])
sns.scatterplot(data = max_human, x = "야간인구수(19~08)", y = max_human.index, ax = ax[1])
plt.show()
```

- 주, 야간 인구수를 비교해봤을 때, 위의 일 최대 이동 인구수 평균과 같은 양상을 보이지만 특이한점은 송파구의 야간인구수가 가장 많이 나타나는 것을 확인 할 수 있다.

```python
gangnam = board_df[board_df["구정보"] == "강남구"]
```

```python
gang_cnt = gangnam.groupby("유형").count()
```

```python
gang_cnt
```

```python
plt.figure()
sns.scatterplot(y = gang_cnt.index, x = gang_cnt["구정보"])
plt.show()
```

- 대표적으로 강남구를 확인해 봤을 때, 보도와 차도가 구분된 도로의 차도에서 수거한 자전거가 가장 많이 등장하였다.
